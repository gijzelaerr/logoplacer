# Finding the optimal spot for a chunk of text 

First we try to find the most salient areas in the image

Secondly we will try to fit a distribution based on the text somewhere in the salient landscape.

http://www.scholarpedia.org/article/Visual_salience#Simple_computational_framework

```python
import cv2
import pyimgsaliency as psal
import PIL.Image
from cStringIO import StringIO
import IPython.display
import numpy as np
```

```python
def showarray(a, fmt='png'):
    a = np.uint8(a)
    f = StringIO()
    PIL.Image.fromarray(a).save(f, fmt)
    IPython.display.display(IPython.display.Image(data=f.getvalue()))
```

```python
# path to the image
filename = '../assets/bird.jpg'

# get the saliency maps using the 3 implemented methods
rbd = psal.get_saliency_rbd(filename).astype('uint8')

ft = psal.get_saliency_ft(filename).astype('uint8')

mbd = psal.get_saliency_mbd(filename).astype('uint8')

# often, it is desirable to have a binary saliency map
binary_sal = psal.binarise_saliency_map(mbd, method='adaptive')

img = cv2.imread(filename)
```

```python
showarray(img)
```

```python
showarray(rbd)
```

```python
showarray(ft)
```
